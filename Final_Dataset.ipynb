```python
import pandas as pd
```

```python
years = range(2006, 2017) # 2006 to 2016
```

```python
world_gdp_df = pd.read_csv("./gdp_from_world_bank/API_NY.GDP.PCAP.CD_DS2_en_csv_v2.csv", skiprows=4)

keep_columns = ['Country Name', 'Country Code'] + map(str, years)
# print(keep_columns)
world_gdp_df = world_gdp_df[keep_columns]

world_gdp_df.head()
```

```python
rename_gdp_dict = {}
for year in map(str, years):
    rename_gdp_dict[year] = ("gdp_%s" % year)
    
world_gdp_df = world_gdp_df.rename(columns=rename_gdp_dict)

world_gdp_df.head()
```
