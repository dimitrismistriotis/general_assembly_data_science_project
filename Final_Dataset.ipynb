```python
import pandas as pd
```

```python
years = range(2011, (2016 + 1)) # 2006 to 2016
```

```python
world_gdp_df_raw = pd.read_csv("./gdp_from_world_bank/API_NY.GDP.PCAP.CD_DS2_en_csv_v2.csv", skiprows=4)

keep_columns = ['Country Name', 'Country Code'] + map(str, years)
# print(keep_columns)
world_gdp_df_raw = world_gdp_df_raw[keep_columns]

world_gdp_df_raw.head()
```

```python
rename_gdp_dict = {
    "Country Name": "country_name",
    "Country Code": "country_code"
}

for year in map(str, years):
    rename_gdp_dict[year] = ("gdp_%s" % year)
    
world_gdp_df = world_gdp_df.rename(columns=rename_gdp_dict)

world_gdp_df.head()
```

## Same for economic freedom

```python
economic_freedom_raw_df = pd.read_csv("./frazer_institute_economic_freedom/economic-freedom-of-the-world-data-for-researchers.csv")
```

```python
economic_freedom_raw_df["Year"].unique()
```

```python
economic_freedom_raw_df[economic_freedom_raw_df["Year"] == years[0]]
```

```python
economic_freedom_partial_dfs = []

for year in years:
    economic_freedom_partial_dfs.append(economic_freedom_raw_df[economic_freedom_raw_df["Year"] == year][
        ["ISO_Code", "Countries", "Economic Freedom Summary Index"]].rename(columns={
            "ISO_Code": "economic_freedom_country_code",
            "Countries": "economic_freedom_country_name",
            "Economic Freedom Summary Index": ("economic_freedom_index_%s" % year)}))


len(economic_freedom_partial_dfs)
```

```python
economic_freedom_df = economic_freedom_partial_dfs[0]
# economic_freedom_df  = economic_freedom_df.reindex(columns=economic_freedom_df.columns)

for i in range(1, len(economic_freedom_partial_dfs)):
    # print(economic_freedom_partial_dfs[i].head())
    current = economic_freedom_partial_dfs[i]
    if not current.empty:
        current = current.rename(columns={
            "economic_freedom_country_code": "economic_freedom_country_code_RIGHT",
            "economic_freedom_country_name": "economic_freedom_country_name_RIGHT",
        })
        # print(current.head())
        economic_freedom_df = pd.merge(economic_freedom_df,
                                       current,
                                       how='inner',
                                       left_on="economic_freedom_country_code",
                                       right_on="economic_freedom_country_code_RIGHT")
        # print(economic_freedom_df.columns)
        del economic_freedom_df['economic_freedom_country_code_RIGHT']
        del economic_freedom_df['economic_freedom_country_name_RIGHT']

economic_freedom_df.head()
```

## Democracy Index from Economist Business Unit

```python
democracy_df = pd.read_csv("./democracy_index_from_wikipedia/democracy_index_2016.csv")

democracy_df.head()
```

```python
democracy_df = democracy_df[["country_name", "score"]].rename(columns={"score": "democracy_score_2016"})

democracy_df.head()
```

```python
from os import listdir
```

```python
import os.path as path
```

```python
indexes_dir = "scrapped_democracy_indexes_from_econominst_bu"

listdir(indexes_dir)
```

```python
democracy_index_years = range(2011, (2015 + 1)) # Used +1 to indicate that the last year is 2015
```

```python
for year in democracy_index_years:
    # print(year)
    target_file_location = path.join(indexes_dir, ("democracy_index_%s.csv" % year))
    # print(target_file_location)
    current_year_df = pd.read_csv(target_file_location)
    current_year_df = current_year_df[["Country", "Overall Score"]].rename(columns={
        "Country": "country_name_RIGHT",
        "Overall Score": ("democracy_score_%s" % year)
    })
    # print(current_year_df.head())
    # Le Mergé:
    democracy_df = pd.merge(democracy_df, current_year_df,
                            how="inner",
                            left_on="country_name",
                            right_on="country_name_RIGHT")
    del democracy_df['country_name_RIGHT']
    # print(democracy_df.head())
```
